```python
import pandas as pd

df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})

# Apply sum column-wise (axis=0)
col_sum = df.apply(sum, axis=0)
print(col_sum)

# Apply sum row-wise (axis=1)
row_sum = df.apply(sum, axis=1)
print(row_sum)
```

```python
# Custom function to subtract the mean from each element
def subtract_mean(x):
    return x - x.mean()

# Apply the custom function column-wise
col_mean_subtracted = df.apply(subtract_mean, axis=0)
print(col_mean_subtracted)
```
