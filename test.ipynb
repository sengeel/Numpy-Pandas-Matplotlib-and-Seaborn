```python
import matplotlib.pyplot as plt
import numpy as np

random_number = np.linspace(0,10, 100)
y1 = np.sin(random_number)
y2 = np.cos(random_number)
y3 = np.tan(random_number)
y4 = np.exp(random_number)

plt.figure(figsize=(10, 10), dpi=100)

plt.subplot(2,2,1)
plt.plot(random_number,y1)
plt.title("Sine Function")

plt.subplot(2,2,2)
plt.plot(random_number,y2)
plt.title("Cosine Function")

plt.subplot(2,2,3)
plt.plot(random_number, y3)
plt.title("Tangent Function")

plt.subplot(2,2,4)
plt.plot(random_number, y4)
plt.title("Exponential Function")

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

datasets = load_dataset('lukebarousse/data_jobs')
df = datasets['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country']=='United States'].copy()
df_US
```

```python
df_US['job_posted_month'] = df['job_posted_date'].dt.strftime('%B')
df_US
```

```python
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_US_pivot
```

```python
df_US_pivot.reset_index(inplace=True)
df_US_pivot
```

```python
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month

df_US_pivot.sort_values(by='month_no', inplace=True)
df_US_pivot.set_index('job_posted_month', inplace=True)
df_US_pivot.drop(columns='month_no', inplace=True)
```

```python
df_US_pivot
```

```python
top_3 = df_US['job_title_short'].value_counts().head(3)
top_3=top_3.index.to_list()

df_US_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings for Top Data Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```

# Explode Functions

```python
import ast
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

datasets = load_dataset('lukebarousse/data_jobs')
df = datasets['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df[['job_title', 'job_skills']].head(5)
```

```python
df_exploded = df.explode('job_skills')
df_exploded
```

```python
skills_count = df_exploded.groupby(['job_title_short', 'job_skills']).size()
skills_count
```

```python
df_skills_count = skills_count.reset_index(name='skill_count')

df_skills_count = df_skills_count.sort_values(by='skill_count', ascending=False)

df_skills_count
```

```python
job_title = 'Data Analyst'
top_skills = 10

df_jobs_count = df_skills_count[df_skills_count['job_title_short'] == job_title].head(top_skills)

df_jobs_count.plot(kind='barh', x='job_skills', y='skill_count', legend='')
plt.gca().invert_yaxis()
plt.title('Top Job Postings Based on Skills')
plt.ylabel("")
plt.xlim(0,100000)
plt.show()
```

# Subplots


```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

datasets = load_dataset('lukebarousse/data_jobs')
df = datasets['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_skills = df.copy()

df_skills = df_skills.explode('job_skills')

skills_count = df_skills.groupby(['job_title_short', 'job_skills']).size()

skills_count
```

```python
df_skills_count = skills_count.reset_index(name='skill_count')
df_skills_count.sort_values(by='skill_count', ascending=False, inplace=True)
df_skills_count
```

```python
job_titles = ['Data Scientist', "Data Engineer" ,"Data Analyst"]
fig, axes = plt.subplots(3,1, figsize=(10, 10), dpi= 100)
for i, job_title in enumerate(job_titles):
    df_plot = df_skills_count[df_skills_count['job_title_short'] == job_title].head(5)
    ax = axes[i]
    df_plot.plot(kind='barh',x='job_skills', y='skill_count', ax=ax, title=job_title, legend="")
    ax.set_xlim(0, 120000)
    ax.set_ylabel('')
    ax.invert_yaxis()
fig.suptitle('Counts of Top Skills in Job Postings', fontsize=15)
fig.tight_layout(rect=[0,0,1,0.96])
```

# Scatter PLots




```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a plot
plt.plot(x, y, marker='x')

# Add text annotations
for i, txt in enumerate(y):
    plt.text(x[i], y[i], f'{txt}', fontsize=12, ha='right')

# Show plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Plot with Annotations')
plt.grid(True)
plt.show()
```
